### Create 2 new columns

#### We need to create two more columns which will be used in further analysis.

```python
# create 2 new columns

epl_df['mins_per_match'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['goals_per_match'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
```

### Total Goals Scored

```python
# total goals

Total_Goals = epl_df['Goals'].sum()
Total_Goals
```

```python
# penalty goals

Total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
# penalty attempts

Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
Total_PenaltyAttempts
```

### Pie Chart

```python
# Pie chart for penalties missed vs penalties scored

# to set a figure size
plt.figure(figsize = (13,6))
pl_not_scored = Total_PenaltyAttempts - Total_PenaltyGoals
data = [pl_not_scored , Total_PenaltyGoals]

# assign labels to the pie chart
labels = ['Penalties missed' , 'Penalties Scored']

# used seaborn library to set color palette
color = sns.color_palette('rocket_r')

# to plot pie chart
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

### Players based on their age group

```python
# players based on their age group

# created different bins of age groups to get exact number players in that age group
under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
above30 = epl_df[epl_df['Age'] > 30]
```

```python
# Used Numpy() library to count total no of players in each bins
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(),
             above30['Name'].count()])

# created labels
mylabels = ('<=20', '>20 & <=25', '>25 & <= 30', '>30' )
# set fig size
plt.figure(figsize = (12,7))
# created title to the pie chart
plt.title('Total players with Age Groups' , fontsize = 20)
# to plot a pie chart
plt.pie(x, labels = mylabels, autopct = '%.2f%%', colors = sns.color_palette('Paired'))
plt.show()
```

```python
# Total under 20 players in each club

# to plot a bar chart to get under 20 players in each club.
players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar' , figsize = (12,5),
color = sns.color_palette('cubehelix'))
```

### Average age of the players in each club

```python
# Avg age of the players in each club

# set a fig size
plt.figure(figsize = (13, 7))

# used seaborn library to plot a boxplot for avg age of the players
# In boxplot, we get clear distribution of numeric values, especially when we want to compare
# them in multiple groups
# This provides high-level information at a glance. Also offers info like min, max, average,
# first quartile, third quartile, variance and any outliers.
sns.boxplot(x = 'Club' , y = 'Age' , data= epl_df )

# sets X-axis tick values, which are the locations along the x-axis where tick marks apperar
plt.xticks(rotation = 90)
```

```python
# avg age of the players grouped by club in tabular form
num_player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum()) / num_player

# sort the data in descending order
data.sort_values(ascending = False)
```

### Total assists from each club

```python
# total assists from each club

# performed groupby() on Club and Assists column 
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club' , as_index=False)['Assists'].sum())

# to set background color and theme
sns.set_theme(style='whitegrid', color_codes= True )

# to vizsualize bar chart 
ax = sns.barplot(x = 'Club' , y = 'Assists', data= Assists_by_clubs.sort_values(by='Assists'),
                 palette = 'Set2')
ax.set_xlabel('Club' , fontsize = 30)  # x- axis
ax.set_ylabel('Assists' , fontsize = 30)  # y - axis
plt.xticks(rotation = 75)   # tick marks along the x-axis
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('Club vs Total Assists' , fontsize = 30)  # to set title
```

```python
# Top 10 Assists

# used groupby() on Assists column 
top_10_assists = epl_df[['Name' , 'Club', 'Assists', 'Matches']].nlargest(n = 10 , columns = 'Assists')
top_10_assists
```

```python
# Goals by each clubs

# used Pandas library, groupby() on Club column and calculated total scored goals
goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index= False)['Goals'].sum())

# used seaborn library to set background theme and color code
sns.set_theme(style ='whitegrid' , color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = goals_by_clubs.sort_values(by = 'Goals'),
                 palette = 'rocket')
ax.set_xlabel('Club', fontsize = 30)   # x-axis label
ax.set_ylabel('Goals', fontsize = 30)  # y-axis label
plt.xticks(rotation = 80)   # tick marks along the x-axis
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Clubs vs Total Goals', fontsize = 30)  # title
```

```python
# Most goals by players

# nlargest() on Goals column and arranged in tabular form
top_10_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n = 10, columns = 'Goals')
top_10_goals
```

```python
# Pie Chart - goals with assists and without assists

plt.figure(figsize = (14, 6))  # fig size
assists  = epl_df['Assists'].sum()
data = [Total_Goals - assists , assists]
labels = ['Goals without assists' , 'Goals with assists ']  # labels of pie chart
color = sns.color_palette('Set1')   # color palette 

# to plot a pie chart
plt.pie(data, labels = labels, colors = color, autopct = "%.0f%%")
plt.show()
```
